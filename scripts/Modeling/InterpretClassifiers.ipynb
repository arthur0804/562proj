```python
import pandas as pd
import numpy as np
import re
import string
import nltk
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import label_ranking_loss
from sklearn.metrics import hamming_loss
from sklearn.metrics import jaccard_similarity_score
import joblib
```

```python
# load the tf-idf vectorizer
tfidf_vectorizer = joblib.load('./saved_models/new_tfidf.pkl')
```

```python
feature_names = tfidf_vectorizer.get_feature_names()
```

```python
# load the classifier
BinaryClassifier = joblib.load('saved_models/BinaryClassifier.pkl')
classifiers = BinaryClassifier.classifiers_
```

```python
tones = ['Amusing', 'Angst-filled', 'Atmospheric', 'Bittersweet', 'Bleak',
       'Chaste', 'Conservative', 'Creepy', 'Darkly humorous',
       'Disturbing', 'Dramatic', 'Emotionally intense', 'Explicit',
       'Feel-good', 'Funny', 'Gossipy', 'Gross', 'Gruesome', 'Haunting',
       'Heartwarming', 'Heartwrenching', 'High-drama', 'Homespun',
       'Hopeful', 'Impartial', 'Impassioned', 'Inspiring', 'Irreverent',
       'Liberal', 'Melancholy', 'Menacing', 'Mildly sensuous', 'Moody',
       'Moving', 'Mystical', 'Noisy', 'Nostalgic', 'Offbeat', 'Patriotic',
       'Quiet', 'Racy humor', 'Reflective', 'Romantic', 'Sad', 'Sardonic',
       'Scary', 'Self-deprecating', 'Serious', 'Silly', 'Sobering',
       'Spiritual', 'Steamy', 'Strong sense of place', 'Suspenseful',
       'Sweet', 'Thought-provoking', 'Upbeat', 'Violent', 'Whimsical']
```

```python
for i in range(0, 59):
    print("{}".format(tones[i]))
    classifier = classifiers[i]
    coefs = classifier.coef_[0]
    dic = {}
    for i in range (0, len(coefs)):
        dic[feature_names[i]] = coefs[i]
    l = sorted(dic.items(), key = lambda x: x[1], reverse=True)
    print("\t.{}".format(l[0:5]))
    print("------------------------------------")
```
